```python
"""
Histopathology Cancer Detection with CNN Architectures
=====================================================
This script builds and compares two CNN architectures (standard and with batch normalization)
for classifying histopathology images for cancer detection.
"""

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.callbacks import ModelCheckpoint, ReduceLROnPlateau
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import seaborn as sns

print("Tensorflow version:", tf.__version__)

# Set random seed for reproducibility
SEED = 42
np.random.seed(SEED)
tf.random.set_seed(SEED)

# Check for GPU availability
physical_devices = tf.config.list_physical_devices('GPU')
print(f"GPUs available: {len(physical_devices)}")
if len(physical_devices) > 0:
    tf.config.experimental.set_memory_growth(physical_devices[0], True)
    print("GPU memory growth enabled")

# For this demo, we'll create a small synthetic dataset of histopathology-like images
def generate_synthetic_histopathology_data(num_samples=500, img_size=32):
    """Generate synthetic histopathology data for demonstration"""

    X = np.zeros((num_samples, img_size, img_size, 3), dtype=np.float32)
    y = np.zeros(num_samples, dtype=np.uint8)

    for i in range(num_samples):
        # Determine if this sample is cancerous (1) or non-cancerous (0)
        is_cancer = np.random.choice([0, 1])
        y[i] = is_cancer

        # Generate base tissue texture
        base_color = np.random.uniform(0.5, 0.8)  # Base pink/purple color for H&E staining
        tissue = np.random.normal(base_color, 0.1, (img_size, img_size, 3))

        # Add cellular structures
        num_cells = np.random.randint(10, 30) if is_cancer else np.random.randint(5, 15)

        for _ in range(num_cells):
            # Random cell location
            x_pos, y_pos = np.random.randint(2, img_size-2, 2)

            # Cell size and color variation based on cancer/normal
            if is_cancer:
                # Cancer cells are typically larger and more irregularly shaped
                cell_size = np.random.randint(2, 6)
                # Darker staining for cancer cells
                cell_color = np.random.uniform(0.2, 0.4)
            else:
                # Normal cells are smaller and more uniform
                cell_size = np.random.randint(1, 3)
                # Lighter staining for normal cells
                cell_color = np.random.uniform(0.3, 0.5)

            # Draw cell
            for dx in range(-cell_size, cell_size+1):
                for dy in range(-cell_size, cell_size+1):
                    if (0 <= x_pos+dx < img_size and 0 <= y_pos+dy < img_size and
                        dx*dx + dy*dy <= cell_size*cell_size):
                        # Cell body (purple-ish in H&E)
                        tissue[x_pos+dx, y_pos+dy, 0] = cell_color  # More red
                        tissue[x_pos+dx, y_pos+dy, 1] = cell_color * 0.7  # Less green
                        tissue[x_pos+dx, y_pos+dy, 2] = cell_color * 1.2  # More blue

        # Add cancer-specific features
        if is_cancer:
            # Add nuclear pleomorphism (variation in nuclear size/shape)
            for _ in range(5):
                x_pos, y_pos = np.random.randint(5, img_size-5, 2)
                irreg_size = np.random.randint(3, 7)
                irreg_color = np.random.uniform(0.1, 0.3)  # Very dark nuclei

                # Draw irregular nucleus
                for dx in range(-irreg_size, irreg_size+1):
                    for dy in range(-irreg_size, irreg_size+1):
                        if (0 <= x_pos+dx < img_size and 0 <= y_pos+dy < img_size and
                            dx*dx + dy*dy <= irreg_size*irreg_size * np.random.uniform(0.7, 1.3)):  # Irregular shape
                            tissue[x_pos+dx, y_pos+dy, :] = irreg_color

        # Clip values to [0, 1] range
        X[i] = np.clip(tissue, 0, 1)

    return X, y

# Generate training and validation data
print("Generating synthetic histopathology data...")
X_train, y_train = generate_synthetic_histopathology_data(800, 32)
X_val, y_val = generate_synthetic_histopathology_data(200, 32)

# Display class distribution
print("Class distribution in training data:")
unique, counts = np.unique(y_train, return_counts=True)
print(dict(zip(unique, counts)))

# Visualize some sample images
plt.figure(figsize=(12, 8))
for i in range(8):
    plt.subplot(2, 4, i+1)
    plt.imshow(X_train[i])
    plt.title(f"Label: {'Cancer' if y_train[i] == 1 else 'Normal'}")
    plt.axis('off')
plt.tight_layout()
plt.show()

# Data augmentation setup
train_datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True,
    vertical_flip=True,
    fill_mode='nearest'
)

val_datagen = ImageDataGenerator()

# Prepare the data generators
batch_size = 32
train_generator = train_datagen.flow(X_train, y_train, batch_size=batch_size)
val_generator = val_datagen.flow(X_val, y_val, batch_size=batch_size, shuffle=False)

# Define the Standard CNN model
def create_standard_model():
    model = models.Sequential([
        # First convolutional block
        layers.Input(shape=(32, 32, 3)),
        layers.Conv2D(64, (3, 3), padding='same'),
        layers.Activation('relu'),
        layers.MaxPooling2D((2, 2)),

        # Second convolutional block
        layers.Conv2D(128, (3, 3), padding='same'),
        layers.Activation('relu'),
        layers.MaxPooling2D((2, 2)),

        # Third convolutional block
        layers.Conv2D(256, (3, 3), padding='same'),
        layers.Activation('relu'),
        layers.MaxPooling2D((2, 2)),

        # Flatten and dense layers
        layers.Flatten(),
        layers.Dense(256),
        layers.Activation('relu'),
        layers.Dropout(0.5),

        # Output layer
        layers.Dense(1, activation='sigmoid')
    ])

    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
        loss='binary_crossentropy',
        metrics=['accuracy', tf.keras.metrics.AUC()]
    )

    return model

# Define the CNN model with Batch Normalization
def create_batchnorm_model():
    model = models.Sequential([
        # First convolutional block with batch normalization
        layers.Input(shape=(32, 32, 3)),
        layers.Conv2D(64, (3, 3), padding='same'),
        layers.BatchNormalization(),
        layers.Activation('relu'),
        layers.MaxPooling2D((2, 2)),

        # Second convolutional block with batch normalization
        layers.Conv2D(128, (3, 3), padding='same'),
        layers.BatchNormalization(),
        layers.Activation('relu'),
        layers.MaxPooling2D((2, 2)),

        # Third convolutional block with batch normalization
        layers.Conv2D(256, (3, 3), padding='same'),
        layers.BatchNormalization(),
        layers.Activation('relu'),
        layers.MaxPooling2D((2, 2)),

        # Flatten and dense layers
        layers.Flatten(),
        layers.Dense(256),
        layers.BatchNormalization(),
        layers.Activation('relu'),
        layers.Dropout(0.5),

        # Output layer
        layers.Dense(1, activation='sigmoid')
    ])

    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
        loss='binary_crossentropy',
        metrics=['accuracy', tf.keras.metrics.AUC()]
    )

    return model

# Create and display the model architectures
standard_model = create_standard_model()
print("\nStandard CNN Architecture:")
standard_model.summary()

batchnorm_model = create_batchnorm_model()
print("\nBatch Normalization CNN Architecture:")
batchnorm_model.summary()

# Callbacks for training
callbacks = [
    ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=3, min_lr=1e-6),
    ModelCheckpoint('best_standard_model.h5', monitor='val_accuracy', save_best_only=True, verbose=1)
]

batchnorm_callbacks = [
    ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=3, min_lr=1e-6),
    ModelCheckpoint('best_batchnorm_model.h5', monitor='val_accuracy', save_best_only=True, verbose=1)
]

# Training parameters
epochs = 10
steps_per_epoch = len(X_train) // batch_size
validation_steps = len(X_val) // batch_size

# Train the standard model
print("\nTraining Standard CNN Model...")
standard_history = standard_model.fit(
    train_generator,
    epochs=epochs,
    steps_per_epoch=steps_per_epoch,
    validation_data=val_generator,
    validation_steps=validation_steps,
    callbacks=callbacks,
    verbose=1
)

# Train the batch normalization model
print("\nTraining Batch Normalization CNN Model...")
batchnorm_history = batchnorm_model.fit(
    train_generator,
    epochs=epochs,
    steps_per_epoch=steps_per_epoch,
    validation_data=val_generator,
    validation_steps=validation_steps,
    callbacks=batchnorm_callbacks,
    verbose=1
)

# Plot training history
plt.figure(figsize=(15, 10))

# Plot training & validation accuracy values
plt.subplot(2, 2, 1)
plt.plot(standard_history.history['accuracy'], label='Standard Train')
plt.plot(standard_history.history['val_accuracy'], label='Standard Val')
plt.plot(batchnorm_history.history['accuracy'], label='BatchNorm Train')
plt.plot(batchnorm_history.history['val_accuracy'], label='BatchNorm Val')
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend()

# Plot training & validation loss values
plt.subplot(2, 2, 2)
plt.plot(standard_history.history['loss'], label='Standard Train')
plt.plot(standard_history.history['val_loss'], label='Standard Val')
plt.plot(batchnorm_history.history['loss'], label='BatchNorm Train')
plt.plot(batchnorm_history.history['val_loss'], label='BatchNorm Val')
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend()

# Plot training & validation AUC values
plt.subplot(2, 2, 3)
if 'auc' in standard_history.history:
    plt.plot(standard_history.history['auc'], label='Standard Train')
if 'val_auc' in standard_history.history:
    plt.plot(standard_history.history['val_auc'], label='Standard Val')
if 'auc' in batchnorm_history.history:
    plt.plot(batchnorm_history.history['auc'], label='BatchNorm Train')
if 'val_auc' in batchnorm_history.history:
    plt.plot(batchnorm_history.history['val_auc'], label='BatchNorm Val')
plt.title('Model AUC')
plt.ylabel('AUC')
plt.xlabel('Epoch')
plt.legend()

plt.tight_layout()
plt.show()

# Evaluate models on validation data
standard_pred = standard_model.predict(X_val)
batchnorm_pred = batchnorm_model.predict(X_val)

standard_pred_binary = (standard_pred > 0.5).astype(int).reshape(-1)
batchnorm_pred_binary = (batchnorm_pred > 0.5).astype(int).reshape(-1)

# Compute confusion matrices
cm_standard = confusion_matrix(y_val, standard_pred_binary)
cm_batchnorm = confusion_matrix(y_val, batchnorm_pred_binary)

# Display confusion matrices
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.heatmap(cm_standard, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Normal', 'Cancer'],
            yticklabels=['Normal', 'Cancer'])
plt.title('Standard CNN Confusion Matrix')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')

plt.subplot(1, 2, 2)
sns.heatmap(cm_batchnorm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Normal', 'Cancer'],
            yticklabels=['Normal', 'Cancer'])
plt.title('BatchNorm CNN Confusion Matrix')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')

plt.tight_layout()
plt.show()

# Print classification reports
print("\nStandard CNN Classification Report:")
print(classification_report(y_val, standard_pred_binary))

print("\nBatchNorm CNN Classification Report:")
print(classification_report(y_val, batchnorm_pred_binary))

# Plot ROC curves
fpr_standard, tpr_standard, _ = roc_curve(y_val, standard_pred)
fpr_batchnorm, tpr_batchnorm, _ = roc_curve(y_val, batchnorm_pred)

auc_standard = auc(fpr_standard, tpr_standard)
auc_batchnorm = auc(fpr_batchnorm, tpr_batchnorm)

plt.figure(figsize=(8, 6))
plt.plot(fpr_standard, tpr_standard, label=f'Standard CNN (AUC = {auc_standard:.3f})')
plt.plot(fpr_batchnorm, tpr_batchnorm, label=f'BatchNorm CNN (AUC = {auc_batchnorm:.3f})')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve Comparison')
plt.legend(loc='lower right')
plt.grid(alpha=0.3)
plt.show()

# Create comparison table
comparison_df = pd.DataFrame({
    'Model': ['Standard CNN', 'BatchNorm CNN'],
    'Validation Accuracy': [
        np.max(standard_history.history['val_accuracy']),
        np.max(batchnorm_history.history['val_accuracy'])
    ],
    'Validation Loss': [
        np.min(standard_history.history['val_loss']),
        np.min(batchnorm_history.history['val_loss'])
    ],
    'Validation AUC': [auc_standard, auc_batchnorm]
})

print("\nModel Performance Comparison:")
print(comparison_df)

# Visualize predictions on some validation samples
plt.figure(figsize=(15, 8))
for i in range(8):
    plt.subplot(2, 4, i+1)
    plt.imshow(X_val[i])
    plt.title(f"True: {'Cancer' if y_val[i] == 1 else 'Normal'}\n" +
              f"Standard: {standard_pred[i][0]:.2f}\n" +
              f"BatchNorm: {batchnorm_pred[i][0]:.2f}")
    plt.axis('off')
plt.tight_layout()
plt.show()

print("Completed!")
```
